```python
### Building a model to predict the possibility of side effects on drug properties
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Drop missing values
df_ml = df.dropna(subset=["side_effects", "drug_classes"])

# Convert side effects into a list of individual effects
df_ml["side_effects_list"] = df_ml["side_effects"].apply(lambda x: x.split("; "))

df_sample = df_ml.sample(n=1000, random_state=42)

# Convert side effects to a multi-label format
mlb = MultiLabelBinarizer()
y_multilabel = mlb.fit_transform(df_sample["side_effects_list"])

# Convert drug classes into a numerical format using TF-IDF
vectorizer = TfidfVectorizer(max_features=100)
X_multilabel = vectorizer.fit_transform(df_sample["drug_classes"].astype(str))

# Check the shape of inputs and outputs
X_multilabel.shape, y_multilabel.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_multilabel, y_multilabel, test_size=0.2, random_state=42)

# Train a multi-label classifier using Random Forest
clf_multi = MultiOutputClassifier(RandomForestClassifier(n_estimators=100, random_state=42))
clf_multi.fit(X_train, y_train)

# Predict on test data
y_pred = clf_multi.predict(X_test)

# Compute F1-score for multi-label classification
f1 = f1_score(y_test, y_pred, average="micro")  # Micro-averaged F1-score
f1
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

# Use Logistic Regression with One-vs-Rest (OvR) for multi-label classification
clf_ovr = OneVsRestClassifier(LogisticRegression(max_iter=500, solver="liblinear"))
clf_ovr.fit(X_train, y_train)

# Predict on test data
y_pred_ovr = clf_ovr.predict(X_test)

# Compute F1-score for multi-label classification
f1_ovr = f1_score(y_test, y_pred_ovr, average="micro")
f1_ovr
```

```python
from sklearn.metrics import f1_score

# Compute F1-score
f1_ovr = f1_score(y_test, y_pred_ovr, average="micro")
print(f"Micro-Averaged F1-Score: {f1_ovr:.4f}")
```

```python
import numpy as np

y_test = np.array(y_test)
y_pred_ovr = np.array(y_pred_ovr)

# Pick a random test sample
index = np.random.randint(0, len(y_test))

# Get actual side effects
actual_effects = mlb.inverse_transform(y_test[index].reshape(1, -1))[0]

# Get predicted side effects
predicted_effects = mlb.inverse_transform(y_pred_ovr[index].reshape(1, -1))[0]

print(f"Actual Side Effects: {actual_effects}")
print(f"Predicted Side Effects: {predicted_effects}")
```
